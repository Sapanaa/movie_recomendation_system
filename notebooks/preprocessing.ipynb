```python
df = df.drop_duplicates().reset_index(drop=True)
```

**Severe class imbalance**

- 99.98% of movies are labeled False
- Only 9 movies are labeled True
- This provides almost no predictive value.

```python

missing_count = df.isnull().sum()

missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing Percentage': missing_percent.round(2)
})

missing_df = missing_df.sort_values(by='Missing Percentage', ascending=False)

missing_df
```

Columns with more than 60% missing values `belongs_to_collection`, `homepage` were removed to reduce sparsity and avoid introducing noise into the recommendation model. Features were retained only if they provided meaningful semantic or behavioral information for similarity computation.

Identifier Columns - `id`, `imdb_id`, These are identifiers and do not help measure similarity.

Financial Columns- `budget`, `revenue`, These do not influence content similarity between movies.

Low Informational Value - `adult`, `video`, `status`.These do not contribute meaningfully to recommendation logic.
