```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/movies.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.dropna(subset=['title'])
```

```python
df.isna().sum()
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
df.isna().sum()
```

```python
df.iloc[0]['genres']
```

```python
df['genres'] = df['genres'].str.findall("'name': '([^']*)'")

df['genres'] = df['genres'].apply(lambda x: " ".join(x))
```

```python
df['genres']
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
df.isnull().sum()
```

```python
df['tags'] = df['overview']+' '+df['genres']+' '+df['tagline']
```

```python
df['tags'].head()
```

```python
df['tags'][0]
```

```python
df.to_csv('data/movies_w_tag.csv')
```
